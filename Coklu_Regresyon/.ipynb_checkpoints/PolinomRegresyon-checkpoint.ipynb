```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler #yeni olarak Polynomial'i getirdik
from sklearn.metrics import r2_score ,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline #yeni olarak bide bunu getirdik
```

```python
data_frame=pd.read_csv("customersatisfaction.csv")
```

```python
data_frame.head()
```

##### 'unnamed' column'u normalde datayi kaydeden kisi index column'unu silmeyi unutursa cikabilir karsimiza. Silelim bunu

```python
data_frame.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
data_frame.head()
```

```python
plt.scatter(data_frame["Customer Satisfaction"], data_frame["Incentive"], color="black")
plt.xlabel("Customer St")
plt.ylabel("Incentive")
plt.show()
```

### goruldugu uzere sanki polinomal sekilde korelasyon oldugunu goruyoruz

### ilk olarak dependent ve independent leri belirleyelim

```python
X=data_frame[["Customer Satisfaction"]]
y=data_frame["Incentive"]
```

### train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

### scaler

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

### regression

```python
regresyon = LinearRegression()
```

```python
regresyon.fit(X_train,  y_train)
```

### prediction

```python
y_tahmin=regresyon.predict(X_test)
```

```python
r2score= r2_score(y_test, y_tahmin)
print(r2score)
```

#### gordugumuz gibi r2 skoru cok dusuk cikti 

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regresyon.predict(X_train), color="black")
plt.show()
```

#### __plt.plot(X_train, regresyon.predict(X_train), color="black")__ tüm X_train değerleri için yaptığı tahminleri birleştiren düz bir çizgi çizdiniz. Bu çizgi, modelin veriden öğrendiği en iyi uyum sağlayan regresyon doğrusudur;

### Degree

```python
poly=PolynomialFeatures(degree=2, include_bias=True) #diyerek objesini olusturalim. Aklina kullanimi ile ilgili bisi gelmezse arastir
```

#### degree'yi deneyerek en iyi modeli buluruz genelde. <br> <br>__include_bias=True__-->Bu, modelinize bir sabit terim (bias), yani kesişim noktası (intercept) eklemenizi sağlar. Bu, regresyon denkleminizdeki b'yi temsil eder.(y=ax+b).<br>Verilerin basina 1 diye bir column ekler

```python
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

#### basina "1" column'u var. bu intercept olsun diye koyuyor. sonra hesaplanacak

```python
p_regresyon = LinearRegression() #bu sekilde olustururuz regresyonu
p_regresyon.fit(X_train_poly, y_train)
```

```python
poly_y_tahmin=p_regresyon.predict(X_test_poly)
poly_r2score=r2_score(y_test, poly_y_tahmin)
print(poly_r2score)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, p_regresyon.predict(X_train_poly), color="black")
plt.show()
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_train, p_regresyon.predict(X_train_poly), color="black")
plt.show()
```

## Pipeline

```python
#poly=PolynomialFeatures(degree=2, include_bias=True)
#yaptigimiz bu islemi bir fonksiyon icinde daha kolay ve profesyonelce yapmamizi saglar
```

### arka arkaya bircok dereceyi denemek icin filan yapilir

```python
def poly_regresyon(degree):
    poly_feat=PolynomialFeatures(degree=degree)
    line_reg=LinearRegression()
    scaler=StandardScaler()
    pipeline=Pipeline([
        ("standart_scaler", scaler),
        ("poly_feat", poly_feat),
        ("line_reg", line_reg)
    ]) #sira sira uygulamak istedigimiz olaylar neyse buraya tek seferde yazabiliriz

    pipeline.fit(X_train, y_train)
    score2=pipeline.score(X_test, y_test)
    print("r2score:",score2)

    y_new_tahmin=pipeline.predict(new_df)
    plt.plot(new_df, y_new_tahmin, color="black", label="new predict")
    plt.scatter(X_train, y_train, label="Training points", color="red")
    plt.scatter(X_test, y_test, label="test")
    plt.legend()
    plt.show()
```

```python
poly_regresyon(1
              )
```

```python
poly_regresyon(2)
```

```python
poly_regresyon(3)
```

```python
poly_regresyon(4) # gibi gibi deneyebilirsin
```

```python
for i in range(1,11):
    poly_regresyon(i)
```
